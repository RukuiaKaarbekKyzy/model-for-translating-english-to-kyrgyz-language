**Очищает текст от пунктуации и стоп-слов для английского и киргизского языков.
Разделяет данные на обучающую и тестовую выборки.
Токенизирует текст, преобразуя его в числовые последовательности.
Сохраняет результаты в файлы для дальнейшего использования в моделях.**

```python
import nltk
from nltk.corpus import stopwords
import pandas as pd
import re
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

# Загрузка стоп-слов для английского и киргизского языков
nltk.download('stopwords')
stop_words_en = set(stopwords.words('english'))
# Пример простого списка стоп-слов для киргизского языка
stop_words_ky = {'жана', 'менен', 'үчүн', 'башка', 'артык', 'эле', 'жок', 'бул', 'алар', 'биз', 'деп', 'бир'}

# Загрузка данных
df = pd.read_csv(r"C:\Users\rukui\Downloads\dataset1.csv")

# Функция для очистки текста
def clean_text(text, stop_words):
    # Проверка типа данных
    if not isinstance(text, str):
        return ''  # Возвращаем пустую строку для нестроковых значений
    # Приведение текста к нижнему регистру
    text = text.lower()
    # Удаление пунктуации
    text = re.sub(r'[^\w\s]', '', text)
    # Удаление стоп-слов
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text

# Обработка пропущенных значений
df['text_en'] = df['text_en'].fillna('')
df['text_ky'] = df['text_ky'].fillna('')

# Применение очистки текста с удалением стоп-слов
df['Clean English Text'] = df['text_en'].apply(lambda x: clean_text(x, stop_words_en))
df['Clean Kyrgyz Text'] = df['text_ky'].apply(lambda x: clean_text(x, stop_words_ky))

# Просмотр очищенных данных
print(df.head())

# Разделение данных на обучающую и тестовую выборки
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Сохранение разделенных данных в CSV
train_df.to_csv('train_data.csv', index=False)
test_df.to_csv('test_data.csv', index=False)

# Токенизация текста
tokenizer_en = Tokenizer()
tokenizer_ky = Tokenizer()

# Обучение токенизаторов
tokenizer_en.fit_on_texts(train_df['Clean English Text'])
tokenizer_ky.fit_on_texts(train_df['Clean Kyrgyz Text'])

# Преобразование текста в последовательности чисел
X_train_en = tokenizer_en.texts_to_sequences(train_df['Clean English Text'])
X_train_ky = tokenizer_ky.texts_to_sequences(train_df['Clean Kyrgyz Text'])

# Преобразование текста в последовательности чисел для тестовых данных
X_test_en = tokenizer_en.texts_to_sequences(test_df['Clean English Text'])
X_test_ky = tokenizer_ky.texts_to_sequences(test_df['Clean Kyrgyz Text'])

# Печать результатов токенизации для проверки
print("Пример токенизации (английский):", X_train_en[:5])
print("Пример токенизации (киргизский):", X_train_ky[:5])
```

**Этот код создаёт модель последовательного машинного перевода (seq2seq) на основе LSTM для перевода с английского на киргизский язык. Кратко о его работе:
Подготовка данных:

Выполняется паддинг (добавление нулей) для всех текстов, чтобы они имели одинаковую длину.
Целевые тексты смещаются: последний токен убирается для входа, а первый токен — для выхода.
Архитектура модели:

Энкодер: Двусторонняя LSTM (Bidirectional LSTM) обрабатывает английский текст и возвращает состояния (h, c), которые используются декодером.
Декодер: Обычная LSTM принимает начальные состояния от энкодера и обучается генерировать последовательности слов на киргизском языке.
Выходной слой: Полносвязный слой с функцией активации softmax предсказывает вероятности для каждого слова из киргизского словаря.

Компиляция модели:

Используется оптимизатор Adam.
Функция потерь — sparse_categorical_crossentropy (для задач многоклассовой классификации).
Обучение модели:

Модель обучается на паре входных (английский текст) и целевых (киргизский текст) данных.
Проверяется точность на тестовых данных после каждой эпохи.
Эта модель позволяет обучить систему для перевода предложений с английского на киргизский **

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Embedding, LSTM, Dense, Input, Bidirectional, Concatenate
from tensorflow.keras.optimizers import Adam

# Параметры
embedding_dim = 256  # Размерность векторного пространства для эмбеддингов
hidden_units = 512   # Количество нейронов в слое LSTM
vocab_size_en = len(tokenizer_en.word_index) + 1  # Размер словаря для английского текста
vocab_size_ky = len(tokenizer_ky.word_index) + 1  # Размер словаря для киргизского текста
max_len_en = 13  # Максимальная длина английских текстов
max_len_ky = 13  # Максимальная длина киргизских текстов

# Паддинг для обучающих данных
X_train_en_pad = pad_sequences(X_train_en, maxlen=max_len_en, padding='post', truncating='post')
X_train_ky_pad = pad_sequences(X_train_ky, maxlen=max_len_ky, padding='post', truncating='post')

# Сдвиг для целевых меток (убираем последний токен для входа и первый токен для выхода)
X_train_ky_shifted = X_train_ky_pad[:, :-1]  # Вход для декодера
X_train_ky_target = X_train_ky_pad[:, 1:]    # Целевые метки

# Паддинг для тестовых данных
X_test_en_pad = pad_sequences(X_test_en, maxlen=max_len_en, padding='post', truncating='post')
X_test_ky_pad = pad_sequences(X_test_ky, maxlen=max_len_ky, padding='post', truncating='post')
X_test_ky_shifted = X_test_ky_pad[:, :-1]
X_test_ky_target = X_test_ky_pad[:, 1:]

# Проверка результата
print("Пример паддинга X_train_en_pad:", X_train_en_pad[0])
print("Пример целевых меток X_train_ky_target:", X_train_ky_target[0])

# Определение входов
input_en = Input(shape=(max_len_en,))  # Вход для английского текста
input_ky = Input(shape=(max_len_ky - 1,))  # Вход для киргизского текста (декодер без последнего токена)

# Эмбеддинг для входных данных
embedding_en = Embedding(input_dim=vocab_size_en, output_dim=embedding_dim, input_length=max_len_en)(input_en)
embedding_ky = Embedding(input_dim=vocab_size_ky, output_dim=embedding_dim, input_length=max_len_ky - 1)(input_ky)

# Энкодер (Bidirectional LSTM) для английского текста
encoder_lstm = Bidirectional(LSTM(hidden_units, return_state=True, dropout=0.3, recurrent_dropout=0.3))

# Применяем слой к вектору embedding_en
encoder_output, forward_h, forward_c, backward_h, backward_c = encoder_lstm(embedding_en)

# Получение состояний энкодера для использования в декодере
# Состояния будут объединены для использования в декодере
encoder_state_h = Concatenate()([forward_h, backward_h])  # Состояние h
encoder_state_c = Concatenate()([forward_c, backward_c])  # Состояние c

# Приведение состояний к нужной размерности для декодера
encoder_state_h = Dense(hidden_units, activation='tanh')(encoder_state_h)
encoder_state_c = Dense(hidden_units, activation='tanh')(encoder_state_c)

encoder_states = [encoder_state_h, encoder_state_c]

# Декодер (LSTM) для киргизского текста, с начальными состояниями из энкодера
decoder_lstm = LSTM(hidden_units, return_sequences=True, return_state=True, dropout=0.3, recurrent_dropout=0.3)
decoder_output, _, _ = decoder_lstm(embedding_ky, initial_state=encoder_states)

# Полносвязный слой для предсказания следующего слова
dense = Dense(vocab_size_ky, activation='softmax')
decoder_output = dense(decoder_output)

# Создание модели
model = Model(inputs=[input_en, input_ky], outputs=decoder_output)

# Компиляция модели
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Структура модели
model.summary()

# Обучение модели
history = model.fit(
    [X_train_en_pad, X_train_ky_shifted],
    X_train_ky_target,
    validation_data=([X_test_en_pad, X_test_ky_shifted], X_test_ky_target),
    batch_size=32,
    epochs=10
)
```

Этот код выполняет следующие действия:

Переводит текст с английского на киргизский с помощью Google Translator.
Преобразует исходный текст в числовую последовательность и выполняет паддинг.
Прогнозирует перевод с использованием обученной модели.
Проверяет, были ли предсказаны нулевые индексы (ошибка модели).
Преобразует предсказанные индексы обратно в текст на киргизском.
Пример: переводит фразу "The weather is very cold" с помощью модели и Google Translator.

```python
from deep_translator import GoogleTranslator
import numpy as np

# Функция перевода с использованием Google Translator
def translate(input_text):
    # Перевод текста с английского на киргизский с использованием Google Translator
    translated_text = GoogleTranslator(source='en', target='ky').translate(input_text)

    # Вывод переведенного текста
    print("Переведенный текст с помощью Google Translator:", translated_text)

    # Преобразование текста в последовательность индексов с помощью tokenizer (если нужно дополнительно обработать)
    input_sequence = tokenizer_en.texts_to_sequences([input_text])
    input_pad = pad_sequences(input_sequence, maxlen=max_len_en, padding='post')

    # Прогнозирование (оставляем как есть, если нужно обработать модель)
    output_sequence = model.predict([input_pad, np.zeros((1, max_len_ky - 1))])

    # Вывод предсказанных индексов для диагностики
    predicted_indices = np.argmax(output_sequence, axis=-1)
    print("Predicted indices:", predicted_indices)

    # Проверка на нулевые индексы (или другие исключения)
    if np.all(predicted_indices == 0):
        print("Ошибка: предсказаны только нулевые индексы. Возможно, модель не обучена должным образом.")
        return "Перевод невозможен"

    # Преобразование предсказанных индексов в текст
    output_text = tokenizer_ky.sequences_to_texts(predicted_indices)
    print("Output text:", output_text)  # Дополнительный вывод для проверки
    return output_text[0]

# Пример перевода
translated_text = translate("The weather is very cold")
print("Переведенный текст:", translated_text)
```
